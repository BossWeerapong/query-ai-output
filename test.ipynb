```python
def get_default_parameters() :
    response = requests.get(os.environ.get("API_GET_PARAMETER_URL"))
    if response.status_code == 200 :
        data = response.json()

        defult_parameter = {}
        for result in data["results"] :
            defult_parameter[result['parameter']] = result['values']
        
        return defult_parameter
    
    else:   
        print("Error:", response.status_code, response.text)
```
